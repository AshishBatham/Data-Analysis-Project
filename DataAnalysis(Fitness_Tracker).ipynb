```python
Fitdata["Id"] = Fitdata["Id"].astype(str)
Fitdata["ActivityDate"] = pd.to_datetime(Fitdata["ActivityDate"], format = "%m/%d/%Y")
```

```python
Fitdata["distance_diff"] = Fitdata["TotalDistance"] - Fitdata["TrackerDistance"]
```

```python
Fitdata.columns = Fitdata.columns.str.lower()
Fitdata.columns
```

```python
import datetime as dt
day_of_week = Fitdata['activitydate'].dt.day_name()
Fitdata['week'] = day_of_week
Fitdata['n_week'] = Fitdata['activitydate'].dt.weekday
Fitdata.head(3)
```

```python
Fitdata.rename(columns = { 'activitydate':'activity_date', 'totalsteps':'total_steps', 'totaldistance':'total_distance',
                          'trackerdistance':'tracker_distance','loggedactivitiesdistance':'logged_activities_distance',
                           'veryactivedistance':'very_active_distance','moderatelyactivedistance':'moderately_active_distance',
                           'lightactivedistance':'light_active_distance', 'sedentaryactivedistance':'sedentary_active_distance',
                           'veryactiveminutes':'very_active_minutes', 'fairlyactiveminutes':'fairly_active_minutes',
                            'lightlyactiveminutes':'lightly_active_minutes', 'sedentaryminutes':'sedentary_minutes'}, inplace = True)
```

```python
id_grp = Fitdata.groupby(['id'])
```

```python
Fitdata.head(3)
```

```python
Fitdata['activity_date'].value_counts()
```

```python
id_avg_step = id_grp['total_steps'].mean().sort_values(ascending = False)
id_avg_step = id_avg_step.to_frame()
id_avg_step
```

```python
id_avg_step['activity_level'] = np.where(id_avg_step <= 6000, 'sedentary',
                                np.where(id_avg_step < 12000, 'active', 'very_active'))

id_activity_level = id_avg_step['activity_level']
id_avg_step
```

```python
Fitdata['activity_level'] = [id_activity_level[c] for c in Fitdata['id']]
```

```python
activity_levels = ['very_active', 'active', 'sedentary']
colors = ['red', 'green', 'blue']
fig, ax = plt.subplots(figsize=(8, 5))
for level, color in zip(activity_levels, colors):
    subset = Fitdata[Fitdata['activity_level'] == level]
    ax.scatter(subset['total_steps'], subset['calories'], color=color, label=level, marker='.', s=20)

ax.set_xlabel('Total Steps', fontsize=12)
ax.set_ylabel('Calories', fontsize=12)
ax.set_title('Total Steps vs Calories Burned by Activity Level', fontsize=14, fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.6)
ax.legend(title='Activity Level')

plt.tight_layout()
plt.show()
```

```python
week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
colors = ['blue', 'green', 'red', 'purple', 'orange', 'cyan', 'pink']
fig, ax = plt.subplots(1, 1, figsize=(8, 5))
day_grp = Fitdata.groupby('week')
avg_daily_steps = day_grp['total_steps'].mean()

ax.bar(avg_daily_steps.index, avg_daily_steps.values, color=colors, zorder=2)
ax.set_xticks(range(len(week)))
ax.set_xticklabels(week)
ax.set_xlabel('Day of the Week', fontsize=12)
ax.set_ylabel('Average Daily Steps', fontsize=12)
ax.grid(True, linestyle='--',alpha=0.6, zorder=1)
ax.set_title('Average Steps per Day', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
very_active_minutes = Fitdata['very_active_minutes'].sum()
fairly_active_minutes = Fitdata['fairly_active_minutes'].sum()
lightly_active_minutes = Fitdata['lightly_active_minutes'].sum()
sedentary_minutes = Fitdata['sedentary_minutes'].sum()

slices = [very_active_minutes, fairly_active_minutes, lightly_active_minutes, sedentary_minutes]
labels = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary']
colors = ['blue', 'green', 'red', 'purple']
explode = (0.1, 0.1, 0.1, 0)

plt.pie(slices, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%')
plt.title('% of Activity in Minutes', fontsize =16, fontweight='bold')
plt.axis('equal')
plt.show()
```

```python
activity_level_counts = Fitdata['activity_level'].value_counts()

plt.figure(figsize=(8, 6))
activity_level_counts.plot(kind='bar', color='red', zorder=2)
plt.title('Activity Level Distribution')
plt.xlabel('Activity Level')
plt.ylabel('Frequency')
plt.grid(True, zorder=1)
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
Fitdata[['very_active_minutes', 'fairly_active_minutes', 'lightly_active_minutes', 'sedentary_minutes']].plot.box()
plt.title('Minutes Spent in Different Activity Levels')
plt.ylabel('Minutes')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Fitdata.groupby('week')['total_steps'].mean(), label='Average Total Steps', marker='o', color='blue')
plt.plot(Fitdata.groupby('week')['calories'].mean(), label='Average Calories Burned', marker='o', color='orange')
plt.title('Week-wise Trends in Total Steps and Calories')
plt.xlabel('Week Number')
plt.ylabel('Average Values')
plt.legend()
plt.grid(True)
plt.show()
```
